```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
class LinearRegressor:
    """ 
    Basic linear regression implementation with support for L2 regularizaion
    """ 
    def __init__(self, L2_reg=0):
        self.a = L2_reg
    
    def fit(self, X, y):
        assert X.shape[0] > 0, "Empty input"
        assert X.shape[0] == y.shape[0], "Unequal number of samples in X and y"
        assert len(y.shape) <= 2
        if len(y.shape) == 2:
            assert y.shape[1] == 1
        else:
            y = y[:, None]
        n, d = np.shape(X)
        X = np.hstack((np.ones((n, 1)), X))
        cov_inv = np.linalg.inv(np.matmul(X.T, X) + self.a*np.identity(d + 1))
        self.w = np.matmul(cov_inv, np.dot(X.T, y))
    
    def predict(self, X):
        assert len(X.shape) == 2
        n, d = np.shape(X)
        assert n > 0, "Empty input"
        assert d+1 == np.shape(self.w)[0], "Expected input to have {} features".format(d)
        X = np.hstack((np.ones((n, 1)), X))
        y_pred = np.dot(X, self.w)
        return y_pred
```

```python
dataset = datasets.load_boston()
X = dataset.data
y = dataset.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
model = LinearRegressor(L2_reg = 0)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mse = np.mean((y_test - y_pred)**2)
print("MSE = {}".format(mse))
```

```python
plt.figure()
plt.plot(y_test, label = 'target')
plt.plot(y_pred, label = 'prediction')
plt.legend()
```

```python
train_data, test_data, train_target, test_target = train_test_split(X, y, test_size = 0.30, random_state = 1)
```

```python
# Sample multiple smaller datasets to analyze bias and variance
D = 10
X_train = []
y_train = []
for i in range(D):
    X_tr, X_te, y_tr, y_te = train_test_split(train_data, train_target, test_size = 0.50, random_state = i)
    X_train.append(X_tr)
    y_train.append(y_tr)
```

```python
L2_reg = [0.001, 0.01, 0.1, 1, 10, 100]
bias_sq = []
var = []
for a in L2_reg:
    models = [LinearRegressor(L2_reg = a) for i in range(D)]
    for i in range(D):
        models[i].fit(X_train[i], y_train[i])
    preds = np.hstack([model.predict(test_data) for model in models])
    avg_preds = np.mean(preds, axis = 1)
    v = np.mean((preds - avg_preds[:, None])**2, axis = (0,1))
#     print('Variance = {:.3f}'.format(v))
    b = np.mean((avg_preds - test_target)**2, axis = 0)
#     print('Bias^2 = {:.3f}'.format(b))
    var.append(v)
    bias_sq.append(b)
```

```python
plt.figure()
plt.semilogx(L2_reg, var, label = 'Variance')
plt.semilogx(L2_reg, bias_sq, label = 'Bias^2')
plt.xlabel('Strength of L2 regularization')
plt.legend()
```
